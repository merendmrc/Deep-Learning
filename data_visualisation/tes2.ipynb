```python
import seaborn as sns
data = sns.load_dataset("titanic")
df = data.copy()
df["who"].value_counts()
```

```python
sns.barplot(x="sex", y="who", data= df.groupby["sex","who"])
```

```python
sns.barplot(x= "who",hue="survived" ,y= "count",data=df.groupby(["who","survived"])["survived"].count().reset_index(name= "count"))
```

```python
sns.scatterplot(x= "age", y= "fare", hue= "sex" ,data= df)
```
