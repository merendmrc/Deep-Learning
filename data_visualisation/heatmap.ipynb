```python
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
df = flights.copy()
```

```python
df = df.pivot(index= "month", columns="year", values= "passengers")
```

```python
sns.heatmap(df, cmap= plt.get_cmap("viridis"))
```

```python
sns.heatmap(df, cmap= plt.get_cmap("viridis").reversed(), annot= True, fmt="d", linewidths=.1)
```
