```python
import seaborn as sbn

diamonds = sbn.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

Caprazlama

```python
(sbn.FacetGrid(df, hue= "cut", height= 5, xlim= (0,10000))).map(sbn.histplot, "price").add_legend()
```

```python
(sbn.FacetGrid(df, hue= "cut", height= 5, xlim= (0,10000))).map(sbn.kdeplot, "price", shade= True).add_legend()
```

```python
sbn.catplot(x= "cut", y = "price", data= df, hue="color", kind="point")
```
