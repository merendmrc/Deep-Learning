```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.shape
```

```python
df.info()
```

```python
sns.lineplot(x= "timepoint", y= "signal", data=df, color=sns.color_palette("Set3")[9])
```

```python
sns.lineplot(x= "timepoint", y= "signal", data=df, hue= "event")
df.groupby(["event","timepoint"])["signal"].describe().T
```

```python
sns.lineplot(x= "timepoint", y= "signal", data=df, hue= "event", style="event")
```

```python
sns.lineplot(x= "timepoint", y= "signal", data=df, hue= "event", style="event", markers=True, dashes= False)
```

```python
#ci -> confidence interval(default= 95)
sns.lineplot(x= "timepoint", y= "signal", data=df, hue= "region", style="event", ci= 50, palette= sns.color_palette("Set1", n_colors=4))
```
