```python
import math
import matplotlib.pyplot as plt
import numpy as np 

def sigmoid(x):
    y = []
    for i in x:
        y.append(1/(1+math.exp(-i)))
    return y 

def tanh(x, derivative = False):
    if derivative==True:
        return 1-(x**2)
    return np.tanh(x)

def relu(x):
    y = []
    for i in x:
        if i<=0:
            y.append(0)
        else:
            y.append(i)
    return y
    
def lrelu(x):
    y= []
    for i in x:
        if i<=0:
            y.append(i/10)
        else:
            y.append(i)
    return y
```

```python
x = np.arange(-3, 3, .1)

sig = sigmoid(x)
tan = tanh(x)
rel = relu(x)
lrel = lrelu(x)
swish = sig*x
```

```python
line1 = plt.plot(x ,sig, label = "sigmoid")
line2 = plt.plot(x , tan, label = "TanH")
line1 = plt.plot(x , rel, label = "ReLu")
line1 = plt.plot(x , lrel, label = "LRelu")
line1 = plt.plot(x , swish, label = "swish")

plt.legend()
```
