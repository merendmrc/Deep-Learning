##Matrix transactions

```python
import numpy as np

def naive_add(x,y):
  assert len(x.shape) == 2
  assert x.shape == y.shape

  x = x.copy()

  for i in range(x.shape[0]):
    for j in range(y.shape[1]):
      x[i,j] += y[i,j]

  return x
```

```python
x = np.array([[7,14,21],[8,16,24],[9,18,27]])
y = np.array([[2,2,2],[3,3,3],[5,5,5]])
print(x.shape)
print(y.shape)
naive_add(x,y)
```

```python
x + y
```

```python
def naive_relu(x):
  assert len(x.shape) == 2
  x = x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[0]):
      x[i,j] = max(x[i,j], 0)

  return x
```

```python
x = np.array([[7,-14,21],[-8,16,24],[-9,18,-27]])
naive_relu(x)
```

```python
z = np.maximum(x, 0)
z
```

###Broadcasting

en
---
Operations between vectors, matrices, or tensors with different shapes.

tr (Yayma)
---
Farklı boyutlardaki vektör, matrix, tensorler arasındaki işlemler

```python
def naive_add_matrix_and_vector(x, y):
  assert len(x.shape) == 2
  assert len(y.shape) == 1
  assert x.shape[1] == y.shape[0]
  x= x.copy()
  for i in range(x.shape[0]):
    for j in range(x.shape[1]):
      x[i, j] +=y[j]
  return x
```

```python
x = np.array([[7,14,21],[8,16,24],[9,18,27]])
y = np.array((2,2,2))
naive_add_matrix_and_vector(x, y)
```
